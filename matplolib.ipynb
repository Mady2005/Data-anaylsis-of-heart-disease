```python

y=np.array([9,8,4,3,4,5,7])
plt.pie(y)
plt.show()
```

```python

y=np.array([9,8,4,3,4,5,7])
AA=(['A','B','C','D','E','F','G'])
CC=['#3C3352','#B000B9','#FFED00','#5EE6EB','#EC5858','#FFAB4C','#16FF00']
bb=[0,0,0,0.4,0,0,0.7]
plt.pie(y,labels=AA,autopct='%2.1f%%',colors=CC,explode=bb)
plt.legend(loc="upper right",bbox_to_anchor=(1,1.2))
plt.show()
```

```python
##bargraph
```

```python

y=np.array([9,8,4,3])
AA=(['A','B','C','D'])
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
plt.bar(AA, y, color=bar_colors,label=bar_labels)
plt.title("sports Watch DATA")
plt.xlabel("pulse")
plt.ylabel("calorie")
plt.legend()
plt.show()
```

```python

y=np.array([9,8,4,3])
AA=(['A','B','C','D'])
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
plt.barh(AA, y, color=bar_colors,label=bar_labels)
plt.title("sports Watch DATA")
plt.xlabel("pulse")
plt.ylabel("calorie")
plt.legend()
plt.show()
```
